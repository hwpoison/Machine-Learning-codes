```python
from perceptron import Perceptron
import numpy as np
```

# Clasificador de tomates
El clasificador de tomates ayudará a identificar si un tomate está a punto o aún es inmaduro tomando de referencia 4 puntos.<br>
* Verdor
* Rojez
* Suavidad
* Consistencia

### Preparando los datos 

```python
input_data = np.array([[0.9, 0.1, 0.7, 0.2],

                      [0.0, 0.9, 0.3, 0.8],

                      [0.2, 1.0, 0.0, 0.7],

                      [1.0, 0.0, 0.8, 0.1],

                      [0.3, 0.4, 0.3, 0.6],

                      [0.5, 0.6, 0.4, 0.4]])
# salidas esperadas
expected_output = np.array([0, 1, 1, 0, 1, 0]).T
```

### Inicializando el perceptron

```python
perceptron = Perceptron()
```

Los **pesos** se inicializan en *0*, el **radio de aprendizaje** a *0.01* y el **sesgo** a *1.0*

```python
perceptron.synapse_weights = [0, 0, 0, 0]
perceptron.learn_rate = 0.01
perceptron.bias = 1.0
```

### Entrenamiento
Le pasamos a la neurona las entradas del dataset y la salida esperada por cada una de ellas. Ella va a repasar tantas veces necesite hasta ajustar los pesos de la forma más optima posible.

```python
print("Pesos antes del entrenamiento:", perceptron.synapse_weights)
perceptron.train(input_data, expected_output)
print("Pesos luego del entrenamiento:", perceptron.synapse_weights)
```

### Testeo
Una vez finalizado el entrenamiento, podremos probar el modelo.

```python
def detect_tomate(salida):
    if(salida):
        return "Tomate Maduro"
    else:
        return "Tomate Inmaduro"
    
print("Inserte las propiedades de su tomate! [valor del 0.0 al 1.0]:")
while True:
    try:
        verdor = float(input("\n\nCantidad de verdor:"))
        rojez = float(input("Cantidad de rojez:"))
        consistencia = float(input("Consistencia:"))
        suavidad = float(input("Suavidad:"))
        salida = detect_tomate(perceptron.predict(
            [verdor, rojez, consistencia, suavidad]))
        print("=> Es un", salida)
    except ValueError as error:
        print(f"Por favor introduce un numero valido. [{error}]")
```
